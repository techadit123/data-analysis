```python
# Load the Titanic dataset from Seaborn
df = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print(df.head())

# Get information about the dataset
print(df.info())

# Get summary statistics of the dataset
print(df.describe())

# Check for missing values
print(df.isnull().sum())

```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot histograms for numerical columns
sns.set()
plt.figure(figsize=(10, 6))
sns.histplot(df['age'].dropna(), bins=20)
plt.title('Age Distribution')
plt.show()
```

```python
# Plot bar charts for categorical columns
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df)
plt.title('Sex Distribution')
plt.show()
```

```python
# Plot boxplots for numerical columns by categorical columns
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='age', data=df)
plt.title('Age Distribution by Pclass')
plt.show()
```

```python
# Plot scatterplots for relationships between numerical columns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', data=df)
plt.title('Relationship between Age and Fare')
plt.show()
```

```python
# Use pairplot to visualize relationships between numerical columns
sns.pairplot(df[['age', 'fare']])
plt.show()
```

```python
# Use heatmap to visualize correlations between numerical columns
corr_matrix = df[['age', 'fare']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
